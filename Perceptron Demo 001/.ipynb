```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv('placement-dataset.csv')
```

```python
print (df.shape)
df.head()
```

```python
df = df.drop(df.columns[0], axis = 1)
```

```python
df.head()
```

```python
sns.scatterplot(x = 'cgpa', y = 'iq', hue = 'placement', data = df)
```

```python
X = df.iloc[ : , 0 : 2]
y = df.iloc[ : , -1]
```

```python
print ("X:\n", X.head())
print ("y:\n", y.head())
```

```python

p = Perceptron()
p.fit(X, y)
```

```python
print ("Weights:", p.coef_)
print ("Bias:", p.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values, y.values, clf = p)
```
