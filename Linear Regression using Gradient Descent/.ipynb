```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_dataset.csv')
df.head()
```

```python
df = df.iloc[ : , 1 : ]
df.head()
```

```python
X = df.iloc[ : , 0]
y = df.iloc[ : , -1]
```

```python
plt.scatter(X, y)
plt.show()
```

```python
learning_rate = 0.01
epochs = 2500
m = len(X)
theta_0, theta_1 = 0, 0
cost_history = []
for _ in range(epochs):
    
    y_pred = theta_0 + theta_1 * X
    d_theta0 = (1 / m) * (np.sum(y_pred - y))
    d_theta1 = (1 / m) * (np.sum(X * (y_pred - y)))

    theta_0 = theta_0 - (learning_rate * d_theta0)
    theta_1 = theta_1 - (learning_rate * d_theta1)

    cost_history.append((1 / (2*m)) * np.sum(np.square(y_pred - y)))

print ("Final Weight Adjustments:")
print (f"y' = {theta_0} + {theta_1}x")

```

```python
y_pred = theta_0 + theta_1*X
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python

# ? Examining cost function
plt.plot(range(epochs), cost_history, color='red')
```
